#### kasumi - 1~9權重

```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import fontManager

# 還沒引用字體
width=0.25

listx=['等級','蔬果','食材','時間','持有上限','主技能內容','主技能等級','能力增加:主技能','能力增加:速度','能力增加:活力','能力增加:無','能力增加:發現','能力減少:主技能','能力減少:速度','能力減少:活力','能力減少:無', '能力減少:發現']
a1=[x-width/2 for x in range(len(listx))]
b1=[x+width/2 for x in range(len(listx))]
a2=[10,20,30,40]
b2=[50,60,70,80]

plt.bar(a1,a2,width,label='a')
plt.bar(b1,b2,width,label='b')
plt.title("kasumi",fontsize='20')
plt.xlabel("加成")
plt.xticks(range(len(listx)),labels=listx)
plt.legend()
plt.show()
```
